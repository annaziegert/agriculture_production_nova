### __plot_quantity_correlation__ method

```python
Group22.plot_quantity_correlation(agros_data, agros_data.my_df)
```

As presented in the correlation matrix, most of the quantities have a high correlation (over 0.85). The high correlations can also be explained that countries try to spread their investments in agriculture equally and not only rely on a single pillar of production. The same holds true for countries that are less developed as they do not have the means to invest significantly in any agricultural sector. 

The lowest correlation is between pasture_quantity and fish_output_quantity (0.55), which can be explained which pasture being a land covered with grass for grazing animals, while fish live in the sea/ocean. The in comparison low correlation between fish_output_quantity and any other quantity-columns is due to the fact that most of the quantity columns are earth-related. Aditionally, the fishing sector is highly dependent on the geographical location of the respective country, which further lowers certain correlations.
